# Data Visualization

### Matplotlib and Seaborn lib used for the Data Visualoization

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# pyplot is a sub-lib of the matplotlib
```

```python
a = list(range(1, 15, 3))

a
```

```python
# Creating list using list comprehension

b = [random.randint(10,100) for i in range(5)]

b
```

## Line Chart

#### Simple Line Chart

```python
plt.plot(a,b)        # .plot(x-axis, y-axis)
```

```python
# Above it gives object info/properties\

# If I want to remove the object info



# Way - 1

plt.plot(a,b);
```

```python
# Way - 2

plt.plot(a,b)

plt.show()
```

#### Giving Title & Label to the Line Chart

```python
plt.plot(a,b)
plt.title("Relationships B/W A & B")
plt.xlabel("A")
plt.ylabel("B");
```

## Saving Visuals as Image

#####  Save the Image in the current folder/path 


```python

plt.plot(a,b)
plt.title("Relationships B/W A & B")
plt.xlabel("A")
plt.ylabel("B")

plt.savefig("A Vs b.jpg")      # Other format - .png, .jpeg 
```

# Formating

## `-->` Legend

```python
# Above we dont know which color line A or B has. 

# To know that we have to give legend to the chart, it means we have to do formating  
```

```python
plt.plot(a,b)
plt.plot(a,c)
plt.title("Relationships B/W A Vs B & C")
plt.xlabel("A")
plt.ylabel("B & C")

plt.legend(labels = ["B","C"]);
```

### If I want to replace the Legend place


It can we located at -

1. Upper Right
2. Upper Left
3. Lower Right
4. Lower Left

By default it stays at Upper Right


```python
plt.plot(a,b)
plt.plot(a,c)
plt.title("Relationships B/W A Vs B & C")
plt.xlabel("A")
plt.ylabel("B & C")

plt.legend(labels = ["B","C"], loc = "lower left");
```

```python
plt.plot(a,b)
plt.plot(a,c)
plt.title("Relationships B/W A Vs B & C")
plt.xlabel("A")
plt.ylabel("B & C")

plt.legend(labels = ["B","C"], loc = "lower right");
```

```python
plt.plot(a,b)
plt.plot(a,c)
plt.title("Relationships B/W A Vs B & C")
plt.xlabel("A")
plt.ylabel("B & C")

plt.legend(labels = ["B","C"], loc = "upper left");
```
