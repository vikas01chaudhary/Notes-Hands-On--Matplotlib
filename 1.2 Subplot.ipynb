# Subplot

--> It just break one chart into multiple pieces
<br>
-->  or make sub-charts of any chart
<br>

--> We can access these sub-charts using indexing like:

        ______0____|_____1_______
    1  |     (0,0) | (0,1)       |
       |___________|_____________|
    1  |      (1,0)| (1,1)       |
       |___________|_____________|
                   |

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import names
```

```python
a = list(range(1, 15, 3))

a
```

```python
b = [random.randint(10,100) for i in range(5)]

b
```

```python
# Subplot

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,10))




# create the 4 charts present in the 2 rows and 2 columns

# fig is the orginal fig(10,10) that is break down in the 2rows & 2 columns = 4 sub-charts

# ax is the axis of every sub chart
```

```python

fig, ax = plt.subplots(nrows = 2, ncols = 4, figsize =(15,10))




# create the 4 charts present in the 2 rows and 4 columns
```

## Ploting with values 

```python

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,10))

# Giving values

ax[0,0].plot(a,'g--')
ax[0,1].plot(b,'k*')
ax[1,0].plot(a,b,'y--o')
ax[1,1].plot(b,a,'D');          # D for daimond shape
```

## Giving Titles to Fig & Sub-Charts

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,10))

# Giving values

ax[0,0].plot(a,'g--')
ax[0,1].plot(b,'k*')
ax[1,0].plot(a,b,'y--o')
ax[1,1].plot(b,a,'D'); 


# Giving values

ax[0,0].set_title('Var A')
ax[0,1].set_title('Var B')
ax[1,0].set_title('A vs B')
ax[1,1].set_title('B vs A')


# Giving Title to entire Chart

fig.suptitle('Buliding the Subplot for the 1st time');
```

# Controlling Limit/Scale of the Subplot/sub-chart

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,10))

# Giving values

ax[0,0].plot(a,'g--')
ax[0,1].plot(b,'k*')
ax[1,0].plot(a,b,'y--o')
ax[1,1].plot(b,a,'D'); 


# Giving values

ax[0,0].set_title('Var A')
ax[0,1].set_title('Var B')
ax[1,0].set_title('A vs B')
ax[1,1].set_title('B vs A')


# Giving Title to entire Chart

fig.suptitle('Buliding the Subplot for the 1st time')



# Controlling the scales of subcharts

ax[0,0].set_ylim(0,15)         # here we have passed only 1 value i.e a (that's why have only ylim)
ax[0,1].set_ylim(0,150)         # here we have passed only 1 value i.e b (that's why have only ylim)

ax[1,0].set_xlim(0,15)
ax[1,0].set_ylim(0,150)

ax[1,1].set_xlim(0,150)
ax[1,1].set_ylim(0,15);
```

# Giving label to subplots

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize =(10,10))

# Giving values

ax[0,0].plot(a,'g--')
ax[0,1].plot(b,'k*')
ax[1,0].plot(a,b,'y--o')
ax[1,1].plot(b,a,'D'); 


# Giving values

ax[0,0].set_title('Var A')
ax[0,1].set_title('Var B')
ax[1,0].set_title('A vs B')
ax[1,1].set_title('B vs A')


# Giving Title to entire Chart

fig.suptitle('Buliding the Subplot for the 1st time')



# Controlling the scales of subcharts

ax[0,0].set_ylim(0,15)         # here we have passed only 1 value i.e a (that's why have only ylim)
ax[0,1].set_ylim(0,150)         # here we have passed only 1 value i.e b (that's why have only ylim)

ax[1,0].set_xlim(0,15)
ax[1,0].set_ylim(0,150)

ax[1,1].set_xlim(0,150)
ax[1,1].set_ylim(0,15);



# Giving labels


ax[0,0].set_xlabel('X-label VAR A')
ax[0,0].set_ylabel('Y-label VAR A')

ax[0,1].set_xlabel('X-label VAR B')
ax[0,1].set_ylabel('Y-label VAR B')

ax[1,0].set_xlabel('X-label VAR A')
ax[1,0].set_ylabel('Y-label VAR B')

ax[1,1].set_xlabel('X-label VAR B')
ax[1,1].set_ylabel('Y-label VAR A');
```
