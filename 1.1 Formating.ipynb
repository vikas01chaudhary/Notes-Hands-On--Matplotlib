# Formating

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import names
import random
```

```python
a = list(range(1, 15, 3))

a
```

```python
# Creating list using list comprehension

b = [random.randint(10,100) for i in range(5)]

b
```

## `-->`   Visual/Fig. Size

```python
plt.figure(figsize = (10,2))               # (figsize = (width, height))
plt.xlabel("Var A")
plt.ylabel("Var B")
plt.plot(a,b);
```

#### Markers :  these are the styles of line

## -, --, --o, --. , *, ^, v, >, <



# Changing the color of line

```python
# WAY 1 -

plt.plot(a,b, color = "Red");
```

```python
# WAY 2 -

plt.plot(a,b,"g");
```

## Changing sytle of line by using Marker

```python
plt.plot(a,b,"g--"); 
```

```python
plt.plot(a,b,"c--o");      # o is not zero, its small alphabatic o


# Here, o is the marker & we can increase or decrease the size of the marker
```

### Increasing the size of the marker

```python


plt.plot(a,b,"c--o", markersize= 10);
```

```python
plt.plot(a,b,"m*");
```

```python
plt.plot(a,b,"y>");
```

```python
plt.plot(a,b,"k^");
```

### Changing Line width

```python
# Increasing the width of the line

plt.plot(a,b,"m--", linewidth = 3);
```

# Limit (Controlling the scale of x-axis & y-axis)

```python

plt.plot(a,b)

plt.xlabel("Var A")
plt.ylabel("Var B")

plt.xlim(0,15)   # (start,end)
plt.ylim(0,110);
```
